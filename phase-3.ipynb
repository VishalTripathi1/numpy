## Advanced Operations with Businedd examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Zoomato sales analysis ====")
print("\n sales data shape", sales_data.shape)
print("\n sales data of first 3 restaurants", sales_data[:3])
```

```python
# total sales per year
print("total",np.sum(sales_data, axis=0))

yearly_total = np.sum(sales_data[:, 1:], axis = 0)
print("yt",yearly_total)
```

```python
# Minimum sales per restaurant
min_sales = np.min(sales_data[:, 1:], axis = 1) # sales_data[row , column], axis = (0 for vertical and 1 for row)
print(min_sales)
```

```python
# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis = 0) 
print(max_sales)
```

```python
# Average Sales per Restarant
avg_sales = np.mean(sales_data[:, 1:], axis = 1)
print(avg_sales)
```

```python
# Cumulative sales 
cumsum = np.cumsum(sales_data[:, 1:], axis = 1) 
print(cumsum)

# if we have to print the graph of this we use matplotlib
plt.figure(figsize=(10, 6), facecolor='#e0f7fa')
plt.plot(np.mean(cumsum, axis = 0))
plt.title("Average Cumulative Sales per year across all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector: ", vector1 * vector2)

print("\n Dot product: ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\n Angle is: ", angle)
```

```python
# Vectorised operatiion

restaruant_types = np.array(['biryani', 'chinese', 'pizza', 'burger', 'cafe'])
vectorised_upper = np.vectorize(str.upper)
print(vectorised_upper(restaruant_types))
```
